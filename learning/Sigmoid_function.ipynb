```python
def sigmoid(x):
    return 1/(1 + np.exp(-x)) 
```

```python
def plot_wx(w1,w2,b):
    x1s = []
    x2s = []
    output = []
    for x1 in range(-6,7):
        for x2 in range(-6,7):
            x1s.append(x1)
            x2s.append(x2)
            output.append(sigmoid((w1 * x1) + (w2 * x2) + b))
    plot_3d_surface(x1s,x2s,output)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
 

def plot_3d_surface(x,y,z):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.plot_trisurf(x,y,z, cmap=plt.cm.viridis, linewidth=0.2)
    plt.show()

    # to Add a color bar which maps values to colors.
    surf=ax.plot_trisurf(x,y,z, cmap=plt.cm.viridis, linewidth=0.2)
    fig.colorbar( surf, shrink=0.5, aspect=5)
    plt.show()

    # Rotate it
    ax.view_init(30, 45)
    plt.show()

    # Other palette
    ax.plot_trisurf(x,y,z, cmap=plt.cm.jet, linewidth=0.01)
    plt.show()
```

```python
plot_wx(10,10,5)
```

```python
interact(plot_wx,w1=(-100,100),w2=(-100,100),b=(-100,100))
```

```python
def plot_wx_return(w1,w2,b):
    x1s = []
    x2s = []
    output = []
    for x1 in range(-6,7):
        for x2 in range(-6,7):
            x1s.append(x1)
            x2s.append(x2)
            output.append(sigmoid((w1 * x1) + (w2 * x2) + b))
    return (x1s,x2s,output)
```

```python
res1 = plot_wx_return(6,0,0)
res2 = plot_wx_return(6,0,20)
```

```python
def build_tower(res1,res2):
    x1s = res1[0]
    x2s = res1[1]
    output = []
    for x,y in zip(res1[2],res2[2]):
        output.append(y-x)
    return x1s,x2s,output
```

```python
x1sl,x2sl,outputl = build_tower(res1,res2)
```

```python
plot_3d_surface(x1sl,x2sl,outputl)
```

```python
res1 = plot_wx_return(0,6,0)
res2 = plot_wx_return(0,6,20)
```

```python
def build_tower(res1,res2):
    x1s = res1[0]
    x2s = res1[1]
    output = []
    for x,y in zip(res1[2],res2[2]):
        output.append(y-x)
    return x1s,x2s,output
```

```python
x1sr,x2sr,outputr = build_tower(res1,res2)
```

```python
plot_3d_surface(x1sr,x2sr,outputr)
```

```python
x1s = x1sl
x1s = x1sr
output = []
output_capped = []
for l,r in zip(outputl,outputr):
    addition = l + r
    output.append(addition)
    output_capped.append(1 if addition < 1 else addition)
```

```python
plot_3d_surface(x1s,x2s,output)
```

```python
plot_3d_surface(x1s,x2s,output_capped)
```
